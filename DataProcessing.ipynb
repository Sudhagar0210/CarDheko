```python
import ast
import pandas as pd

# Sample data
data_Col3 = df_chennai["new_car_feature"]

# Function to convert string to dictionary and extract key-value pairs
def extract_specs(specs_str):
    try:
        specs_dict = ast.literal_eval(specs_str)
        top_data = specs_dict['top']
        top_dict = {item['value']: item['value'] for item in top_data}
        data_section = specs_dict['data']
        for section in data_section:
            for item in section['list']:
                top_dict[item['value']] = item['value']
        return top_dict
    except Exception as e:
        print(f"Error processing entry: {e}")
        return {}

# Apply the function to each entry in the Series
specs_data = data_Col3.apply(extract_specs)

# Create a DataFrame
df_chennai_col3 = pd.DataFrame(list(specs_data))

# Display the DataFrame
df_chennai_col3
```
