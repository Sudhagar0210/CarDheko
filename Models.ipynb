```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score


df['price(Lakh)'] = pd.to_numeric(df['price(Lakh)'], errors='coerce')

# Clean 'Engine Displacement' by removing 'cc' and converting to numeric
df['Engine Displacement'] = df['Engine Displacement'].astype(str).str.replace('cc', '').astype(float)
df['Seats'] = df['Seats'].astype(str).str.replace('Seats', '').astype(float)

# Encoding categorical variables
label_enc = LabelEncoder()
df['bt'] = label_enc.fit_transform(df['bt'])
df['transmission'] = label_enc.fit_transform(df['transmission'])
df['owner'] = label_enc.fit_transform(df['owner'])
df['model'] = label_enc.fit_transform(df['model'])
df['Fuel Type'] = label_enc.fit_transform(df['Fuel Type'])
df['Insurance Validity'] = label_enc.fit_transform(df['Insurance Validity'])
df['City'] = label_enc.fit_transform(df['City'])

# Drop rows where target 'price(Lakh)' is NaN
df = df.dropna(subset=['price(Lakh)'])

# Selecting features and target variable
# x = df[['bt', 'transmission', 'owner', 'modelYear', 'Engine Displacement', 'Fuel Type', 'Seats', 'Mileage', 'km']]
# y = df['price(Lakh)']

x = df.drop('price(Lakh)', axis=1)
y = df['price(Lakh)']
```

```python
x
```

```python
# Model 1

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model_liner = LinearRegression().fit(x_train, y_train)
y_pred_liner = model_liner.predict(x_test)

mae = mean_absolute_error(y_test, y_pred_liner)
mse = mean_squared_error(y_test, y_pred_liner)
r2 = r2_score(y_test, y_pred_liner)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```

```python
# Model 2

from sklearn.ensemble import RandomForestRegressor

model_random = RandomForestRegressor().fit(x_train,y_train)
y_pred_random = model_random.predict(x_test)

mae = mean_absolute_error(y_test,y_pred_random)
mse = mean_squared_error(y_test,y_pred_random)
r2 = r2_score(y_test,y_pred_random)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```

```python
# model 3

import xgboost as xgb

model_xgb = xgb.XGBRegressor().fit(x_train,y_train)
y_pred_xgb = model_xgb.predict(x_test)

mae = mean_absolute_error(y_test,y_pred_xgb)
mse = mean_squared_error(y_test,y_pred_xgb)
r2 = r2_score(y_test,y_pred_xgb)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```

```python
# model 4

from sklearn.ensemble import HistGradientBoostingRegressor

model_hist = HistGradientBoostingRegressor().fit(x_train,y_train)
y_pred_hist = model_hist.predict(x_test)

mae = mean_absolute_error(y_test,y_pred_hist)
mse = mean_squared_error(y_test,y_pred_hist)
r2 = r2_score(y_test,y_pred_hist)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```

```python
# Model 5

from sklearn.ensemble import GradientBoostingRegressor

model_gradient = GradientBoostingRegressor().fit(x_train,y_train)
y_pred_grad = model_gradient.predict(x_test)

mae = mean_absolute_error(y_test,y_pred_grad)
mse = mean_squared_error(y_test,y_pred_grad)
r2 = r2_score(y_test,y_pred_grad)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```

```python
# Model 6

from catboost import CatBoostRegressor

model_cat = CatBoostRegressor().fit(x_train,y_train)
y_pred_cat = model_cat.predict(x_test)

mae = mean_absolute_error(y_test,y_pred_cat)
mse = mean_squared_error(y_test,y_pred_cat)
r2 = r2_score(y_test,y_pred_cat)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```

```python
# Model 7

from sklearn.ensemble import AdaBoostRegressor

model_ada = AdaBoostRegressor().fit(x_train,y_train)
y_pred_ada = model_ada.predict(x_test)

mae = mean_absolute_error(y_test,y_pred_ada)
mse = mean_squared_error(y_test,y_pred_ada)
r2 = r2_score(y_test,y_pred_ada)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```
